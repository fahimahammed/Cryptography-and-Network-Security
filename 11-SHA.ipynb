```python
import hashlib

def encrypt_this_string(input_string):
    # Create a new sha1 hash object
    sha1 = hashlib.sha1()

    # Update the hash object with the input string encoded to bytes
    sha1.update(input_string.encode('utf-8'))

    # Get the hexadecimal digest of the hash
    hashtext = sha1.hexdigest()

    return hashtext

# Driver code
if __name__ == "__main__":
    print("HashCode Generated by SHA-1 for:")

    s1 = "Fahim Ahammed firoz"
    print(f"\n{s1} : {encrypt_this_string(s1)}")

    s2 = "hello world"
    print(f"\n{s2} : {encrypt_this_string(s2)}")
```
